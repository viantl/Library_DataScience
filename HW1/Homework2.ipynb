```python
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
x = np.linspace(-3, 3, 51)
x
```

```python
# Создайте массивы y1, y2, y3, y4 по следующим формулам: y1 = x**2    y2 = 2 * x + 0.5   y3 = -3 * x - 1.5   y4 = sin(x)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и 
# массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк 
# и двух столбцов. 
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

# В каждом графике массив x используется для координат по горизонтали.
# В левом верхнем графике для координат по вертикали используйте y1, в правом верхнем - y2, в левом нижнем - y3, 
# в правом нижнем - y4.
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

# Дайте название графикам: 'График y1', 'График y2' и т.д.
ax1.set_title('График y1')
ax2.set_title('График y2')
ax3.set_title('График y3')
ax4.set_title('График y4')

# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
ax1.set_xlim([-5, 5])

# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
fig.set_size_inches(8, 6)

# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
# Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class 
t=df['Class'].value_counts()
t
```

```python
# На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций 
# (Class равен 1) и другую - для обычных (Class равен 0). 
v1_class1=df.set_index('Class')['V1'].filter(like='1', axis=0)
v1_class1=v1_class1.reset_index()
v1_class1=v1_class1.drop('Class', axis=1)

v1_class0=df.set_index('Class')['V1'].filter(like='0', axis=0)
v1_class0=v1_class0.reset_index()
v1_class0=v1_class0.drop('Class', axis=1)
# Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, 
# а плотность распределения. 
# Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, 
# чтобы гистограммы были полупрозрачными и не загораживали друг друга. 
# Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. 
plt.hist(v1_class0['V1'], bins=20, color='grey', edgecolor='black', density = True, orientation = 'horizontal',alpha=0.5)
plt.hist(v1_class1['V1'], bins=20, color='red', edgecolor='black', density = True, orientation = 'horizontal',alpha=0.5)
# Создайте легенду с двумя значениями: “Class 0” и “Class 1”. 
plt.legend(labels=['Class 0', 'Class 1'])
# Горизонтальной оси дайте название “V1”.
plt.xlabel('V1')
plt.plot()
```

```python
# 11. Создайте одномерный массив c, составленный из последовательности 16-ти случайных равномерно распределенных 
# целых чисел от 0 до 16 невключительно. 
c=np.random.randint(0, 16,16)
c
```

```python
# 12. Поменяйте его форму так, чтобы получилась квадратная матрица C. 
C=c.reshape(4,4)
print(f'C = \n{C}')
# Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10. 
D=B+C*10
print(f'D = \n{D}')
# Вычислите определитель, ранг и обратную матрицу D_inv для D.
print(f'Определитель D: {np.linalg.det(D)}')
print(f'Ранг D: {np.linalg.matrix_rank(D)}')
D_inv=np.linalg.inv(D)
print(f'D_inv:\n {D_inv}')
```

```python
# 13. Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице. 
# Убедитесь, что в матрице D_inv остались только нули и единицы. 
print(f'D_inv:\n{D_inv}')
D_inv=np.where(D_inv<0, 0,1)
print(f'D_inv:\n{D_inv}')
# С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных, 
# получите матрицу E размером 4x4.  
# Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть равны соответствующему 
# элементу матрицы B, 
# а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0, 
# должны быть равны соответствующему элементу матрицы C.
print(f'B:\n{B}')
print(f'C:\n{C}')
E=np.where(D_inv==1, B, C)
print(f'E:\n{E}')
```
